```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal
```

```python
# Generate samples
sample=normal(loc=50,scale=5, size=1000)
```

```python
# Plot the histogram
plt.hist(sample, bins=10)
```

```python
# Calculate sample mean and sample standard deviation
sample_mean=sample.mean()
sample_SD=sample.std()
```

```python
# Fit distribution in above patameters
from scipy.stats import norm
```

```python
dist=norm(sample_mean, sample_SD)
```

```python
# Identify minimum value and maximum value
values=np.linspace(sample.min(), sample.max(),100)
```

```python
# Here we are generating 100 datapoints in minimum and maximum line space 
# and putting probability destribution function.
prob_density=[]
for value in values:
    A=dist.pdf(value)
    prob_density.append(A)
```

```python
# Plotting the histogram for 100 generated values
plt.hist(sample, bins=10, density=True)
plt.plot(values, prob_density)
```

```python
# Also we can get histogram by seaborn library
import seaborn as sns
sns.distplot(sample)
```
